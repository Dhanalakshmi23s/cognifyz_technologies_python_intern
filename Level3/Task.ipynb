### Data Visualization Tool

```python
#import matplotlib, pandas library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#read the csv file
population=pd.read_csv('population.csv')

#give the x and y values and label (give color to the bars)
bars=plt.bar(population.Year,population.Population,label='Population in year',color='#4747d1')

#give title for the graph
plt.title('Population Year Wise',fontdict={'fontname':'Algerian','fontsize':20})

#label the x values (give fontname and size for the label)
plt.xlabel('year',fontdict={'fontname':'Arial','fontsize':'15'})

#label the y values (give fontname and size for the label)
plt.ylabel('Population',fontdict={'fontname':'Arial','fontsize':'15'})

#legend the graph
plt.legend()

#show the graph
plt.show()
```
