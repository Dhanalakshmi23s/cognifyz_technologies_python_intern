```python
import re
from collections import Counter

def count_word_occurrences(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            # Read the content of the file
            content = file.read()

            # Use regular expression to extract words
            words = re.findall(r'\b\w+\b', content.lower())

            # Count the occurrences of each word
            word_counts = Counter(words)

            return word_counts

    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
        return None

def display_word_counts(word_counts):
    if word_counts is not None:
        # Sort the words alphabetically
        sorted_words = sorted(word_counts.keys())

        # Display the results
        for word in sorted_words:
            count = word_counts[word]
            print(f"{word}: {count}")

# Get user input for the file path
file_path = input("Enter the path of the text file: ")

# Count word occurrences and display the results
word_counts = count_word_occurrences(file_path)
display_word_counts(word_counts)
```
